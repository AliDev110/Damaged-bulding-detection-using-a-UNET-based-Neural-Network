```python
from osgeo import gdal, osr

# Load the raster image
input_raster_path = "/path/to/your/satellite_image.tif"
dataset = gdal.Open(input_raster_path, gdal.GA_Update)

if not dataset:
    print("Failed to load the raster image.")
    exit(1)

# Define Ground Control Points (GCPs)
gcps = [
    gdal.GCP(10.0, 20.0, 0, 100, 200),  # Example GCP (map_x, map_y, map_z, pixel_x, pixel_y)
    gdal.GCP(15.0, 25.0, 0, 150, 250),  # Add more GCPs as needed
    # ...
]

# Set the GCPs to the dataset
dataset.SetGCPs(gcps, osr.SRS_WKT_WGS84)

# Get the geotransform and projection
geotransform = dataset.GetGeoTransform()
projection = dataset.GetProjection()

# Save the georeferenced image
output_raster_path = "E:\\FAST NUCES\\FYP\\Datasets\\Massachusetts Dataset\\tiff\\test\\23429020_15.tiff"
driver = gdal.GetDriverByName('GTiff')
output_dataset = driver.CreateCopy(output_raster_path, dataset, 0)

if not output_dataset:
    print("Failed to save the georeferenced image.")
    exit(1)

# Apply the transformation and save the georeferenced image
output_dataset.SetGeoTransform(geotransform)
output_dataset.SetProjection(projection)
output_dataset.FlushCache()

print(f"Georeferenced image saved to {output_raster_path}")

# Clean up
dataset = None
output_dataset = None
```
